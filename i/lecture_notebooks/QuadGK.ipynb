```julia
using QuadGK
```

Gauss-Kronrod method, adaptive Gaussian quadrature.

# Example
$$
\int_0^1 \sqrt{x}dx = \frac{2}{3}
$$
Recall that this function is not smooth at the origin. 

```julia
quadgk(x -> sqrt(x), 0, 1)
```

This returns an estimate of the integral and an estimate of the (absolute) error.

```julia
quadgk(x -> sqrt(x), 0, 1, rtol=1e-12) # we can improve the precision by changing the rtol parameter
```

# Example
$$
\int_0^1\frac{dx}{\sqrt{x}} = 2
$$
This has a left endpoint singularity.

```julia
quadgk(x -> 1/sqrt(x), 0, 1)
```

```julia
quadgk(x -> 1/sqrt(x), 0, 1, rtol=1e-12)
```

## Example
Singularity at Infinity
$$
\int_1^\infty \frac{1}{x^2}\sin(x)dx = 0.5040670619069284
$$
Use `Inf` to denote infinty in Julia:

```julia
quadgk(x ->sin(x)/x^2, 1, Inf)
```
