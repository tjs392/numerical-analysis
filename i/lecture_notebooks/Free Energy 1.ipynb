```julia
using QuadGK
using Plots
```

```julia
V(x) = (x^2-1)^2;
xx = LinRange(-1.5, 1.5, 200);
plot(xx, V.(xx), xlabel="x", ylabel="V(x)", title="Double-Well Potential", legend=false)
```

Consider the free energy defined in terms of **inverse temperature**, $\beta$,
$$
F(\beta) = -\beta^{-1} \log Z(\beta) =  -\beta^{-1} \log \int \exp(-\beta V(x))dx
$$

```julia
function free_energy()
    integral, _ = quadgk(t -> exp(- * V(t)), -Inf, Inf)
    return -log(integral)/
end
```

```julia
_values = LinRange(0.1, 5.0, 50);
F_values = [free_energy() for  in _values];
```
