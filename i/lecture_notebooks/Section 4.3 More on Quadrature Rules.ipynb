# Exercise 19
Check exactness of 
$$
\int_{-1}^1 f(x)dx = f\left(-\frac{\sqrt{3}}{3}\right) + f\left(\frac{\sqrt{3}}{3}\right)
$$

```julia
@show sqrt(3)/3;
```

We can empirically check exactness by finding the highest degree $k$ for which $x^k$ is exactly integrated

```julia
k_vals = [0, 1, 2, 3, 4];
x = [-sqrt(3)/3,  sqrt(3)/3];
for k in k_vals
    f = x-> x^k;
    ∫f_exact = ((1)^(k+1) - (-1)^(k+1))/(k+1);
    ∫f_apx=   f(x[1]) + f(x[2]) ;
    @show k;
    println("Error = $(abs(∫f_exact - ∫f_apx))");
end
```

So the degree of precision is 3.  Compare trapezoidal rule to this for:
